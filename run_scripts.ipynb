Analyse article start and end years

```python
import requests
from datetime import datetime

def get_date_years():
    url = "http://admin:123456@localhost:5984/rinr-2026/_find"
    
    payload = {
        "selector": {"type": "article"},
        "fields": ["dc:date"]
    }
    
    response = requests.post(url, json=payload)

    docs = response.json()['docs']
    
    years = []
    for doc in docs:
        date_str = doc.get('dc:date')  
        if date_str:
            year = int(date_str.split('/')[0])
            years.append(year)
    
    if years:
        return {
            "min_year": min(years),
            "max_year": max(years),
            "total_records": len(years)
        }
    return {"min_year": None, "max_year": None}

# Run it
result = get_date_years()
print(f"Min year: {result['min_year']}")
print(f"Max year: {result['max_year']}")
```
